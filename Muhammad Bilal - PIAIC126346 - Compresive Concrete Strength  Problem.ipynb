```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import models, layers, optimizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('compresive_strength_concrete.csv')
```

```python
data.columns = ['Cement', 'Blast Slag', 'Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age (Day)', 'Concrete Compressive Strength']
```

```python
print(data.isnull().sum())
```

```python
x = data.drop(['Concrete Compressive Strength'], axis =1)
y = data[['Concrete Compressive Strength']]
```

```python
train_data , test_data , train_targets , test_targets = train_test_split(x, y, train_size = 0.7, test_size = 0.3, random_state = 42)
```

```python
mean = train_data.mean(axis=0)
train_data -= mean                                          
std = train_data.std(axis=0)                            #Normalizing the data
train_data /= std

test_data -= mean
test_data /= std
```

```python
new_train_data , val_data , new_train_targets , val_targets = train_test_split(train_data, train_targets, test_size = 0.2, random_state = 42)
```

```python
print(new_train_data.shape)
print(val_data.shape)
print(test_data.shape)
```

```python
model = models.Sequential()
model.add(layers.Dense(10, activation='relu', input_shape = (new_train_data.shape[1],)))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(6, activation='relu'))
model.add(layers.Dense(1))
```

```python
model.compile(optimizer = 'rmsprop',
              loss = 'mse',
              metrics = ['mae'])
```

```python
Result = model.fit(new_train_data, new_train_targets, epochs = 100, validation_data = (val_data, val_targets))
```

```python
test_mse_score , test_mae_score = model.evaluate(test_data, test_targets)
```

```python
Prediction = model.predict(test_data)
```
